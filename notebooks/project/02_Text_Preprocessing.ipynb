```python
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
stop_words = set(stopwords.words("english"))

lemmatizer = WordNetLemmatizer()
```

```python
df = pd.read_csv("../../data/raw/Resume.csv")
```

```python
print(df["Resume_str"][0])
```

```python
def clean_resume(text):
    """
    Clean resume text for machine learning.
    """

    # Convert to lowercase
    text = text.lower()

    # Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove punctuation and special characters
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Remove stopwords
    words = text.split()

    words = [word for word in words if word not in stop_words]

    # Lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
sample_resume = df["Resume_str"][0]

print("Original Resume:\n")
print(sample_resume)

print("\n" + "="*80 + "\n")

cleaned_resume = clean_resume(sample_resume)

print("Cleaned Resume:\n")
print(cleaned_resume)
```

```python
df["Cleaned_Resume"] = df["Resume_str"].apply(clean_resume)
```
